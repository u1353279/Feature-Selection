```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import StratifiedKFold
```

```
df=pd.read_csv('/content/drive/My Drive/santandar.csv/train.csv')
```

```
df.head()
```

```
#training and testing data
X=df.drop(columns=['ID','TARGET'],axis=1)
#test=df.drop(columns=['ID_code'],axis=1)
y=df['TARGET']
```

```
def get_correlation(data,threshold):
  corr_col=set()
  corrmat=data.corr()
  for i in range(len(corrmat.columns)):
    for j in range(i):
      if abs(corrmat.iloc[i,j])>threshold:
        colname=corrmat.columns[i]
        corr_col.add(colname)
  return corr_col
```

```
corr_features=get_correlation(X,0.85)
```

```
X_new=X.drop(labels=corr_features,axis=1)
```

```
X_new.shape
```

```
rf=RandomForestClassifier(n_estimators=100)
from sklearn.metrics import accuracy_score
```

```
accuracy=[]

skf=StratifiedKFold(n_splits=10,random_state=786)
skf.get_n_splits(X_new,y)

for train_index,test_index in skf.split(X_new,y):
  X1_train,X1_test=X_new.iloc[train_index],X_new.iloc[test_index]
  y1_train,y1_test=y.iloc[train_index],y.iloc[test_index]

  rf.fit(X1_train,y1_train)
  ypred=rf.predict(X1_test)
  score=accuracy_score(ypred,y1_test)
  accuracy.append(score)
```

```
print(accuracy)
```
